```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import math
import seaborn as sns
```

```python
arr= np.log([0.000076, 0.000084, 0.00208, 0.07645, 2.96, 54.46, 1175.62])
arr2= np.log([0.000109,0.000079,0.000193,0.00151,0.0225,0.326,7.35])
arr3= np.log([0.000107,0.000136,0.000349,0.00117,0.00803,0.01975,0.1723,2.78])

x=[0,1,2,3,4,5,6,7]
labels=[2,4,8,12,16,20,24,31]

sns.set(rc={'figure.figsize':(8,4)})
sns.set_style("darkgrid", {"axes.facecolor": ".97", 'text.color': ".19"})
sns.lineplot(data=arr, label="ucs")
sns.lineplot(data=arr2,label="misplaced")
sns.lineplot(data=arr3,label= "manhattan")
plt.xticks(x, labels)
plt.title("Logarithmic Execution time for different search", fontdict={'fontsize': 15})
plt.xlabel("Depth of Solution", fontdict={'fontsize': 13})
plt.ylabel("Execution time (seconds)", fontdict={'fontsize': 13})
plt.legend(loc=2,prop={'size': 13})
plt.show()
```
